```python
import numpy as np
import pandas as pd
import requests
from datetime import date
```

```python
df = pd.read_csv('used_cars.csv')
r = requests.get('https://economia.awesomeapi.com.br/last/USD-BRL')

dollar = [r.json()['USDBRL']['bid'], r.json()['USDBRL']['create_date']]
ano = date.today().year
milha = 1.609
```

```python
df["price"] = df["price"].str.replace('$','')
df["price"] = df["price"].str.replace(',','')

df["milage"] = df["milage"].str.replace('mi.','')
df["milage"] = df["milage"].str.replace(',','')
```

```python
df["price"] = pd.to_numeric(df["price"], errors= 'coerce')
df["milage"] = pd.to_numeric(df["milage"], errors= 'coerce')
```

```python
df.info()
```

```python
df['price_converted'] = round(df['price'] * float(dollar[0]), 3)
df['cotacao'] = float(dollar[0])
df['atualizacao_cotacao'] = dollar[1]
df['idade'] = ano - df['model_year']
df['kilometers'] = df['milage'] * milha
```
